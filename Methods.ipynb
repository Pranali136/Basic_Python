List Methods
* append(), extend(), remove(), pop(), insert(), clear(), sort()

```python

q=[1,2,3,4]
print(id(q))
q.append([13,12])
print(q)
#output:-139495990550912
#[1, 2, 3, 4, [13, 12]]

q.append('raj','madhu')
print(q)
#op-TypeError: list.append() takes exactly one argument (2 given)

q.append(('raj',1,2))
print(q)
#op:-[1, 2, 3, 4, ('raj', 1, 2)]

q.append([1,2],1)
print(q)
#TypeError: list.append() takes exactly one argument (2 given)

q.append('raj')
print(q)
#op:- [1, 2, 3, 4, ('raj', 1, 2), 'raj']

q.append('1,2,3')
print(q)
#op:- [1, 2, 3, 4, ('raj', 1, 2), 'raj', '1,2,3']

p='raj'
q.append(p)
print(q)
#op: - [1, 2, 3, 4, ('raj', 1, 2), 'raj', '1,2,3', 'raj']

q.extend([1,2,3])
print(q)
#op: - [1, 2, 3, 4, ('raj', 1, 2), 'raj', '1,2,3', 'raj', 1, 2, 3]

q.remove(1)
print(q)
q.pop(4)
print(q)
```

```python
m=[1,2,3]
m.insert(3,'two')
print(m)
m.clear()
print(m)

```

```python
s=[3,4,1,2]
s.sort()
print(s)
```

```python
y=[3,4,1,2]
y.reverse()
print(y)
y.sort(reverse=True)
print(y)
```

Set methods
* union(), intersection(), difference(), add(), remove(), discard()

```python

a={1,2,3,4}
b={3,4,5,6}
print(a.union(b))
print(b.union(a))

print(a.difference(b))
print(b.difference(a))

print(a.intersection(b))
print(b.intersection(a))

print(a.symmetric_difference(b))
print(b.symmetric_difference(a))

print(id(a))
a.add(4)
print(a)
print(id(a))

a.remove(2)
print(a)

a.discard(4)
print(a)






# {1, 2, 3, 4, 5, 6}
# {1, 2, 3, 4, 5, 6}
# {1, 2}
# {5, 6}
# {3, 4}
# {3, 4}
# {1, 2, 5, 6}
# {1, 2, 5, 6}
# 135124925068928
# {1, 2, 3, 4}
# 135124925068928
# {1, 3, 4}
# {1, 3}
```
