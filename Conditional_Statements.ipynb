#####If–Elif–Else Ladder

```python
a=int(input("enter a number:"))
b=int(input("enter a number:"))
sym=str(input("enter a symbol(+,-,/,//,*,%):"))

print(f"{a}+{b}={a+b}")
print(f"{a}-{b}={a-b}")
print(f"{a}*{b}={a*b}")
print(f"{a}/{b}={a/b}")
print(f"{a}//{b}={a//b}")
print(f"{a}%{b}={a%b}")
```

```python
# CALCULATOR USING IF-ELIF
a=int(input("enter a number:"))
b=int(input("enter a number:"))
sy=str(input("enter a symbol(+,-,*,/,//,%)"))

if sy=='+':
  print(f"{a}+{b}={a+b}")
elif sy=='-':
  print(f"{a}-{b}={a-b}")
elif sy=='*':
  print(f"{a}*{b}={a*b}")
elif sy=='/':
  print(f"{a}/{b}={a/b}")
elif sy=='//':
  print(f"{a}//{b}={a//b}")
elif sy=='%':
  print(f"{a}%{b}={a%b}")
else:
  print("invalid symbol")
```

```python
# SALARY BONUS CALCULATION
a=int(input("enter a salary:"))

if a>=55000:
  print(f"{a}:\n Bonus 25% : {a/100*25} \n Total salary {a+(a/100*25)}")
elif a>=40000 and a<56000:
  print(f"{a}:{a}\n Bonus 20% : {a/100*20} \n Total salary {a+(a/100*20)}")
elif a>=30000 and a<=40000:
  print(f"{a}:{a}\n Bonus 15% : {a/100*15} \n Total salary {a+(a/100*15)}")
elif a>=20000 and a<=30000:
  print(f"{a}:{a}\n Bonus 12% : {a/100*12} \n Total salary {a+(a/100*12)}")
elif a>=8000 and a<=20000:
  print(f"{a}:{a}\n Bonus 10% : {a/100*10} \n Total salary {a+(a/100*10)}")
elif a<8000:
  print(f"{a}:{a}\n Bonus 5% : {a/100*5} \n Total salary {a+(a/100*5)}")
else:
  print("invalid salary")
```

####Nested if-else

```python
# NESTED IF-ELSE (MARKS)
a=int(input("enter a mark:"))
if a<=100:
   if a<=80:
      if a<=70:
        if a<=60:
          if a<=50:
            if a<35:
              if a<0:
                print("invalid marks")
              else:
                print("f")
            else:
              print("D")
          else:
            print("C")
        else:
          print("B")
      else:
        print("A")
   else:
    print("A+")
else:
  print("invalid marks")
```
