```python
#67
# Top half
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=' ')
    # Spaces in the middle
    for s in range(2 * (5 - i)):
        print(' ', end=' ')
    # Descending numbers (skip the last number for the middle row)
    start = i
    if i == 5:  # middle row: start from i-1 to avoid duplicate 5
        start = i - 1
    for j in range(start, 0, -1):
        print(j, end=' ')
    print()  # new line

# Bottom half
for i in range(4, 0, -1):
    # Ascending numbers
    for j in range(1, i + 1):
        print(j, end=' ')
    # Spaces in the middle
    for s in range(2 * (5 - i)):
        print(' ', end=' ')
    # Descending numbers
    for j in range(i, 0, -1):
        print(j, end=' ')
    print()  # new line


```
