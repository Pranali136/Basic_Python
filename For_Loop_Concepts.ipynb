```python
#BASIC FOR LOOP WITH RANGE
for i in range(1,6,1):  # Loop starts from 1 and ends at 5
  print(i)               # Prints current value of i
```

```python
# EVEN NUMBERS WITH STEP VALUE
for i in range(0,51,2):  # Loop runs from 0 to 50, jumping by 2
  print(i-1,end=" ")     # Prints i-1 on the same line
```

```python
# REVERSE LOOP WITH NEGATIVE STEP
for i in range(20,0,-2):   # Loop runs backward from 20 to 2
  print(i,end=" ")
```

```python
# CHANGING VARIABLE INSIDE LOOP
a=5
for i in range(1,a):
  print(a)     # Prints current value of a
  a+=1          # Increases a during loop
print(a)      # Prints final value of a
```

```python
# REVERSE COUNTING
a=5
for i in range(a,-5,-1):   # Counts down from 5 to -4
  print(i,end=" ")
```

```python
# LOOP THROUGH DICTIONARY
f={'a':1,'b':2}
for i in f:    # Iterates over dictionary keys
  print(i)

for i in range(len(f)):  # Loop runs length of dictionary
  print(f)                # Prints full dictionary each time
```

```python
# FLOAT VALUE LOOP
a=1.123
for i in range(1,5):
  print(a)
  a-=0.23  # Decreases float value
print(a)
```

```python
# LOOP THROUGH LIST
a=[1,2,3,4,5]
for i in a:
  print(i)
```

```python
# LOOP THROUGH SET
x={1,6,6,1,2}  # Duplicate values are removed in sets

for i in x:
  print(i)
```

```python
# STRING LOOP
R='Rajendra'
for i in range(len(R)):
  print(R)   # Prints full string multiple times
```

```python
# ASCII VALUE EXAMPLES

a=1
for i in range(a,5):
  print(chr(65))
```

```python
a=1
for i in range(a,5):
  print(chr(64+a),end=" ")     # Converts number to alphabet

  a+=1
```

```python
# BASIC STAR PRINTING
for i in range(1,5):
  print('*')

for i in range(1,5):
  print('*',end=" ")
```

# ALPHABET PATTERNS

```python
for i in range(1,5):
  print(chr(65),end=" ")
```

```python
for i in range(1,5):
  print(chr(64+i),end="  ")
```

```python
for i in range(1,5):
  for j in range(1,5):
    print(chr(64+j),end=" ")
  print()
```

```python
for i in range (1,4):
   for j in range(1,4):
    print(chr(87+j),end=" ")
   print()
```

```python
for i in range (1,5):
   for j in range(1,i+1):
    print(chr(96+j),end=" ")
   print()
```

```python
# REVERSE ALPHABET PYRAMID
for i in range(1, 5):
    for j in range(1, i):
        print(" ", end=" ")
    for k in range(100, 96 + i - 1, -1):
        print(chr(k), end=" ")
    print()
```

```python
# DIAMOND STAR PATTERN
# Upper part of diamond
for i in range(1, 5):
    for j in range(4, i, -1):
        print(" ", end=" ")
    for k in range(1, i * 2 + 1):
        print("*", end=" ")
    print()

# Lower part of diamond
for i in range(4, 0, -1):
    for j in range(4, i, -1):
        print(" ", end=" ")
    for k in range(1, i * 2 + 1):
        print("*", end=" ")
    print()
```
