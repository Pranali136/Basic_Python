#ARITHMETIC OPERATORS

* +  : adds two operands
* -  : subtracts second operand from first
* *  : multiplies two operands
* /  : divides first operand by second
* %  : returns remainder  
* // : returns floor division
* ** : returns power




```python

 a=20
 b=2
 print(a+b)
 print(a-b)
 print(a*b)
 print(a/b)
 print(a%b)
 print(a//b)
 print(a*b)
```

#IDENTITY OPERATORS

* is     : returns True if both variables refer to same object
* is not : returns True if variables refer to different objects

```python
a = [1, 2, 3]
b = a

print(a is b)      # True â†’ both refer to the same object
print(a is not b)  # False
```
