ID FUNCTION (MEMORY LOCATION)

* The id() function is used to get the unique identity (memory address) of an object in Python.
* Every object created in Python is stored in memory, and id() helps us identify where that object is stored.
* If two variables have the same value and same id(), they refer to the same object.
* If the value changes for an immutable object (like int, float, string, tuple), a new memory location is created.
* For mutable objects (like list, set, dictionary), the memory location usually remains the same even if the value changes.

```python
print(id(23)) # WE USE ID FUNCTION TO ACCESS VARIABLES MEMORY LOCATION
```

```python
print(id(1.2))
```

```python
print(id('raj'))
```

```python
print(id((1,2,3)))
```

```python
print(id(['r',1]))
```

Variable Swapping and Memory Behavior using id()

```python
a = 10                 # Assign integer value 10 to variable 'a'
print(a)               # Print the value stored in 'a'
print(id(a))           # Print memory location (id) of value 10

b = 20                 # Assign integer value 20 to variable 'b'
print(b)               # Print the value stored in 'b'
print(id(b))           # Print memory location (id) of value 20

a = a + b              # Add values of 'a' and 'b' and store result in 'a'
print(a)               # Print updated value of 'a' (used for swapping logic)

b = a - b              # Subtract old 'b' from 'a' to get original value of 'a'
print(b)               # Print new value of 'b' after swap step

a = a - b              # Subtract new 'b' from 'a' to get original value of 'b'
print(a)               # Print final swapped value of 'a'

print(a)               # Print swapped value of 'a'
print(id(a))            # Print new memory location of 'a' (int is immutable)

print(b)               # Print swapped value of 'b'
print(id(b))            # Print new memory location of 'b'
```

Variable Reassignment and Dynamic Typing

```python
a = 10                 # Assign integer value 10 to variable 'a'
print(a)               # Print the value stored in 'a'
print(id(a))           # Print memory location of integer object 10

b = 'raj'              # Assign string value 'raj' to variable 'b'
print(b)               # Print the string stored in 'b'

c = a                  # Assign value of 'a' to 'c'
                       # Now 'a' and 'c' refer to the same integer object
print(c)               # Print value stored in 'c'

a = b                  # Reassign 'a' to reference the string object 'raj'
print(a)               # Print new value of 'a' (shows dynamic typing)

b = c                  # Reassign 'b' to reference the integer object stored in 'c'
print(b)               # Print new value of 'b'

print(a)               # Print final value of 'a'
print(id(a))           # Print memory location of 'a' after reassignment

print(b)               # Print final value of 'b'
```

Multiple Variable Assignment

```python
a, b, c = 10, 20, 'raj'   # Multiple assignment:
                          # assigns 10 to 'a', 20 to 'b', and 'raj' to 'c' in one line

print(a)                 # Prints the value stored in variable 'a'
print(b)                 # Prints the value stored in variable 'b'
print(c)                 # Prints the value stored in variable 'c'
```
