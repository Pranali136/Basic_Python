```python
# FUNCTION OVERWRITING
def greet(Name,Age):   # This function overwrites the previous greet() function
  print(f"hello {Name}, your age is {Age}")
greet('raj', 21)
```

```python
# PRIME NUMBER FUNCTION
def prime(n):
  c=0   # Counter to track divisibility

  for i in range(2,n):    # Loop checks divisibility from 2 to n-1
    if n%i==0:
      c+=1
      print(f"{n} is not prime")
      break
  if c==0:
    print(f"{n} is prime")
```

```python
prime(125)
```

```python
#*ARGS (VARIABLE NUMBER OF ARGUMENTS)
def raj(*item):    # *item collects all arguments into a tuple
  return item

```

```python
raj(1,6,9,3)
```
