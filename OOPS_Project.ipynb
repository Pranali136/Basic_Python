Employee Management

```python
class Employee(): # Defining Employee class
  empcount=0     # Class variable to count total employees
  incmt=1.10     # Class variable for salary increment rate

  # Constructor runs when object is created
  def __init__(self,first,last,salary):
    self.first=first   # Instance variable for first name
    self.last=last     # Instance variable for last name
    self.salary=salary # Instance variable for salary
    self.email= first.lower()+"."+last.lower()+"@quastech.com"
    Employee.empcount +=1  # Increment employee count

  def fullname(self):  # Method to return full name
   return f"{self.first.title()} {self.last.title()}"

  def increment(self): # Method to increase salary
    self.salary=int(self.salary*Employee.incmt)
```

```python
emp1=Employee("raj","desai",30000) # Creating employee object
emp2=Employee("rajni","kenjale",35000)
emp3=Employee("sagar","gole",10000)
```

```python
emp2.email # Accessing email of emp2
emp2.incmt # Accessing class variable using object
```

```python
# INHERITANCE (DEVELOPER CLASS)

class Developer(Employee):  # Developer inherits from Employee
  def __init__(self, first, last, salary,prog_lang):  # Constructor of child class
    super().__init__(first, last, salary)     # Calling parent constructor
    self.prog_lang=prog_lang    # Programming language of developer
```

```python
dev1=Developer("raj","desai",35000,"python")
dev2=Developer("kiran","yadav",50000,"SQL")
dev3=Developer("rani","kadam",60000,"R")
```

```python
dev1.fullname() # Calling inherited method
dev1.email     # Inherited property
#dev1.prog_lang  # Child class property
```

```python
# MANAGER CLASS
class Manager(Employee):  # Manager inherits Employee
  def __init__(self,first,last,salary,employees=None):
    super().__init__(first,last,salary)  # Calling parent constructor

    if (employees is None):
      self.employees=[]   # Empty employee list
    else:
      self.employees=employees   # Assign provided employee list

  def addemp(self,emp):  # Adds employee if not present
    if emp not in self.employees:
      self.employees.append(emp)

  def removemp(self,emp): # Removes employee if present
    if emp in self.employees:
      self.employees.remove(emp)

  def print_emps(self):  # Prints all employees under manager
    for emp in self.employees:
      print("-->",emp.fullname())
```

```python
mg1=Manager("tanu","pawar",125000,[emp1,emp2])
```

```python
mg1.first # Access manager first name
mg1.email # Access manager email
mg1.addemp(emp3) # Adding new employee
```

```python
mg1.print_emps() # Printing managed employees
```
