 FOR LOOP

```python
a=1
# Variable 'a' is initialized with value 1
for i in range(1,9): # for-loop runs from 1 to 8 (9 is excluded)
  print(i)           # Prints current value of i
  a+=1               # Increments a by 1 each iteration

```

BASIC WHILE LOOP

```python
a=1           # Initialize variable
while a<=5:   # Loop runs as long as condition is True
  print(a)    # Prints current value of a
  a+=1         # Increments a to avoid infinite loop
```

```python
a=1
while a<=5:
  print(a,end=" ")   # Prints numbers on the same line separated by space
  a+=1
print(a)

```

```python
#WHILE LOOP
a=1
while a<=4:     # chr(65) = 'A', so 64 + a converts numbers to letters
  print(chr(64+a),end=" ")
  a+=1
```

```python
# NESTED WHILE LOOP (PATTERN)
x=1
# Outer loop counter
while x<=4:
  a=1           # Inner loop counter resets each time
  while a<=4:
    print(chr(64+a),end=" ")
    a+=1
  print()  # Moves cursor to next line after inner loop
  x+=1
```

# ARITHMETIC OPERATORS

```python
print(123/10)  # Division operator (/) returns float result
print(123//10) # Floor division (//) removes decimal part
print(123%10)  # Modulus (%) gives remainder
1%10           # This line does nothing because result is not printed or stored
```

```python
# REVERSE A NUMBER USING WHILE LOOP
rev=0         # Variable to store reversed number
a=123         # Number to reverse
a1=a          # Store original value for comparison
while a!=0:
  rem=a%10    # Extract last digit
  rev=(rev*10)+rem     # Build reversed number
  a=a//10      # Remove last digit
print(a1)      # Prints original number
print(rev)     # Prints reversed number

```

```python
# PALINDROME CHECK
rev=0
a=121
a1=a
while a!=0:
  rem=a%10
  rev=(rev*10)+rem
  a=a//10
if rev==a1:     # Compare original and reversed number
  print("palindrome")
else:
  print("not palindrome")
```

```python
n = int(input("Enter number: "))   # Takes input from user and converts to intege
temp = n                           # Stores original value
sum = 0                            # Variable to store sum of cubes

for i in str(n):                   # Converts number to string to iterate digit by digit
    sum = sum + int(i)**3           # Cube of each digit added to sum

if sum == temp:
    print("Armstrong number")
else:
    print("Not Armstrong")

# 3️⃣ int(i)**3

# i is currently a string, e.g., '1'.

# int(i) converts it back to an integer: 1

# **3 cubes it: 1**3 = 1

# 4️⃣ sum = sum + int(i)**3

# Adds the cube of the current digit to the running total sum.

# For 153:

# First iteration: sum = 0 + 1**3 = 1

# Second iteration: sum = 1 + 5**3 = 1 + 125 = 126

# Third iteration: sum = 126 + 3**3 = 126 + 27 = 153
```

```python
# ARMSTRONG NUMBERS IN A RANGE
n = int(input("Enter upper limit: "))

for num in range(1, n + 1):
    sum = 0
    temp = num
    for i in str(num):
        sum = sum + int(i)**3
    if sum == temp:
        print(num)
```

```python
#FIBONACCI SERIES
n = int(input("Enter how many terms: "))
a = 0
b = 1
print(a)
print(b)

for i in range(2, n):
    c = a + b # Next Fibonacci number
    print(c)
    a = b      # Shift values
    b = c
```

```python
# PALINDROME USING WHILE LOOP (USER INPUT)
n = int(input("Enter number: "))
temp = n
rev = 0

while n > 0:
    d = n % 10          # Extract last digit
    rev = rev * 10 + d  # Build reverse number
    n = n // 10         # Remove last digit

if temp == rev:
    print("Palindrome number")
else:
    print("Not Palindrome")
```
